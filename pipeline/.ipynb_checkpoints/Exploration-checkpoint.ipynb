```python
def cleantext(x):
    c = x.replace("\n", " ") # remove the lines
    c = re.sub('\[\*\*.*?\*\*\]', '', c) # remove the identifiers and dates
    #c = c.lower() # remove capital letters
    #c = c.translate(str.maketrans("","", string.punctuation)) # remove punctuation
    return c
```

# Name Entity Recognition

Found here https://github.com/AnthonyMRios/pymetamap

and before that need to get authorization -> takes 3 days

```python
from pymetamap import MetaMap
```

```python
mm = MetaMap.get_instance('/Users/Guillaume/public_mm/bin/metamap16')
```

```python
%time concepts, error = mm.extract_concepts([df.TEXT[0]])
```

```python
concepts
```

```python
def metamap_ner(text):
    concepts, error = mm.extract_concepts([text])
    l = [w[4] for w in concepts] # extract the CUI
    return l
```

```python
%time metamap_ner(df.TEXT[0])
```

```python
for i in range(10):
    %time df.TEXT[i:i+1].apply(lambda x: metamap_ner(x))
```
